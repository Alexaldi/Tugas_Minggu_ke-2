```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("cleaned_student_mental_health_depression.csv")

# Pakai cuma fitur Age dan label Depression
df_small = df[["Age", "Depression"]].copy()

X = df_small[["Age"]]          # fitur
y = df_small["Depression"]   
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,      # 80% train, 20% test
    random_state=42,
    stratify=y          # biar proporsi kelas seimbang
)

print("Jumlah data latih :", len(X_train))
print("Jumlah data uji   :", len(X_test))
print("-" * 60)
```

```python
rf_all = RandomForestClassifier(
    n_estimators=5,
    criterion="gini",
    max_depth=1,
    max_features=1,
    bootstrap=True,
    random_state=42
)

rf_all.fit(X_train, y_train)
```

```python
y_pred_train = rf_all.predict(X_train)
y_pred_test = rf_all.predict(X_test)

print("Akurasi TRAIN :", accuracy_score(y_train, y_pred_train))
print("Akurasi TEST  :", accuracy_score(y_test, y_pred_test))
print("\n=== CLASSIFICATION REPORT (TEST) ===")
print(classification_report(y_test, y_pred_test, digits=3))

print("\nContoh beberapa prediksi di data uji:")
result_sample = pd.DataFrame({
    "Age": X_test["Age"].values,
    "Depression_Actual": y_test.values,
    "Depression_Pred": y_pred_test
}).head(15)
print(result_sample)
```
