```python
import numpy as np
import pandas as pd
from math import exp

# Load dataset
df = pd.read_csv("cleaned_student_mental_health_depression.csv")

# Ambil 6 data pertama (5 train, 1 test)
train = df.iloc[:5]
test = df.iloc[5:6]

# Pilih fitur: Age, Year, CGPA
X_train = train[['Age', 'Year', 'CGPA']].values
y_train = train['Depression'].values
X_test = test[['Age', 'Year', 'CGPA']].values
```

```python
# Bobot awal (bisa disesuaikan)
w = np.array([0.8, -0.1, 0.2])
b = 0.4
eta = 0.1  # learning rate

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# === EPOCH 1 ===
z = np.dot(X_train, w) + b
y_pred = sigmoid(z)
error = y_pred - y_train

dw = np.dot(X_train.T, error) / len(y_train)
db = np.mean(error)

# Update bobot
w = w - eta * dw
b = b - eta * db

loss = -np.mean(y_train*np.log(y_pred + 1e-9) + (1 - y_train)*np.log(1 - y_pred + 1e-9))
print("Epoch 1:")
print(f"Loss = {loss:.6f}")
print(f"w = {w}")
print(f"b = {b:.6f}")
```

```python
# === EPOCH 2 ===
z = np.dot(X_train, w) + b
y_pred = sigmoid(z)
error = y_pred - y_train

dw = np.dot(X_train.T, error) / len(y_train)
db = np.mean(error)

# Update bobot
w = w - eta * dw
b = b - eta * db

loss = -np.mean(y_train*np.log(y_pred + 1e-9) + (1 - y_train)*np.log(1 - y_pred + 1e-9))
print("Epoch 2:")
print(f"Loss = {loss:.6f}")
print(f"w = {w}")
print(f"b = {b:.6f}")
```

```python
# === Pengujian ===
z_test = np.dot(X_test, w) + b
p1 = sigmoid(z_test)[0]
p0 = 1 - p1
y_pred_class = 1 if p1 >= 0.5 else 0

print("=== HASIL PENGUJIAN ===")
print(f"z = {z_test[0]:.6f}")
print(f"P(y=1|x) = {p1:.6f}")
print(f"P(y=0|x) = {p0:.6f}")
print(f"Prediksi kelas = {y_pred_class}")
```
