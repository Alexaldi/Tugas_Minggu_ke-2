```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# Data Latih
X_train = np.array([
    [0, 18, 1, 3],
    [1, 21, 2, 3],
    [1, 19, 1, 3],
    [0, 22, 3, 3],
    [1, 23, 4, 3],
    [1, 19, 2, 3.5],
    [0, 23, 2, 3.5],
    [0, 18, 1, 3.5],
    [0, 19, 2, 2.5],
    [1, 18, 1, 3.5]
])

# Labels latih
y_train = np.array([
    1, 
    0, 
    1, 
    1,  
    0, 
    0, 
    1,  
    0,  
    0,  
    0   
])

# Data Uji
X_test = np.array([
    [0, 20, 1, 3.5]     
])
```

```python
# KLASIFIKASI KNN UNTUK 11 DATA
for k in [1, 3, 5, 7]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    print(f"K = {k} --> Prediksi kelas: {pred[0]}")
```
